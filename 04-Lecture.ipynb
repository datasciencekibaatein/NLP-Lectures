```python
from sklearn.preprocessing import OneHotEncoder
import numpy as np
vocab = np.array(['The','food','is','good','bad','Pizza','amazing']).reshape(-1,1)
vocab
```

```python
encoder = OneHotEncoder(sparse_output=False)

one_hot_matrix = encoder.fit_transform(vocab)
```

```python
for word,vector in zip(vocab,one_hot_matrix):
    print(f"{word[0]}:{vector}")
```

```python
from sklearn.feature_extraction.text import CountVectorizer

import pandas as pd

reviews = [
    'The food is good',
    'The food is bad',
    'Pizza is amazing'
]


vectorizer = CountVectorizer(binary=True)

X = vectorizer.fit_transform(reviews)

matrix_df = pd.DataFrame(X.toarray(),columns=vectorizer.get_feature_names_out())

print(matrix_df)
```

```python
vocab = sorted(list(set(" ".join(reviews).lower().split())))
```

```python
vocab
```

```python
word_to_idx = {word: i for i, word in enumerate(vocab)}
```

```python
word_to_idx
```

```python
for i , review in enumerate(reviews,1):
    sentence_matrix = []
    tokens = review.lower().split()

    for word in tokens:
        vector = [0] * len(vocab)
        vector[word_to_idx[word]] = 1
        sentence_matrix.append(vector)

    print(f"D{i}:{sentence_matrix}")
```
