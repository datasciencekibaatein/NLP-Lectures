<h2><strong>Building blocks of NLP</strong></h2>

<h3>1. Corpus(paragraph)</h3>

A corpus is collection of text data/paragraph

corpus = I am data science student, today learning NLP. I also work with MNC.

<h3>2. Document (sentences)</h3>

A document is a single unit of text inside a corpus.

- doc1 = I am data science student , today learning NLP.
- doc2 = I also work with MNC.

<h3>3. Vocabulary(unique words)</h3>

Unique word present in the corpus.

I like to take coffee. And My Friend like to take tea.


vocab = 9

- No repetitions
- Case sensitive

<h3>4. Words / Tokens</h3>

<h2><strong>Stemming</strong></h2>

Stemming is the process of reducing a word to its root/base form by removing suffixes.

<h4>Port Stemmer</h4>

```python
from nltk.stem import PorterStemmer

ps = PorterStemmer()

words = ['playing',"played","plays","studies","history","classification","running"]

for word in words:
    print(word+"--->"+ps.stem(word))
```

<h4>RegexpStemmer</h4>

It is type of stemmer that uses regular expressions (RegEx) to remove suffixes from words.

```python
from nltk.stem import RegexpStemmer

stemmer = RegexpStemmer('ing$|s$|able$|ed$')

words = ["plays","playing","played","going","gone","defined","moveable"]

for word in words:
    print(word+"---->"+stemmer.stem(word))
```

<h4>Snowball Stemmer</h4>

```python
from nltk.stem import SnowballStemmer

stemmer = SnowballStemmer("english")

words = ['playing',"played","plays","studies","history","classification","running"]

for word in words:
    print(word+"--->"+stemmer.stem(word))
```
