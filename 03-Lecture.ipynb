<h1><strong>Named entity Recognition(NER)</strong></h1>

NER is a task of NLP used to identify and classify important entities in text into predefined category

<h3><strong>These categories usually include:</strong></h3>

- Person names
- Organizations
- Locations
- Dates
- Times
- Money
- Percentages
- Products
- Events
- Nationalities
- Domain specific terms

<h3><strong>Example:</strong></h3> Elon musk founded Tesla in California in 2000.

<h3><strong>Why NER is important</strong></h3>

NER helps machine understands who, what, where & when in text.


- Bussiness
- Healthcare
- Finance
- News
- Search Engines

<h3><strong>Types of Named Entites</strong></h3>


|Label | Meaning |
|------|-------|
|Person|Human name|
|ORG|Organization|
|GPE|Country/city|
|LOC|General Location|
|Date|date|
|Time|Time|
|Money|Currency|
|Product|Product|
|EVENT|Event|

```python
import spacy

nlp = spacy.load("en_core_web_sm")

text = "Elon Musk visited India and met Prime Minister Narendra Modi."

doc = nlp(text)

for ent in doc.ents:
    print(ent.text,":",ent.label_)
```
