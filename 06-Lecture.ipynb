<h1><strong>TF-IDF</strong></h1>

TF-IDF is a statistical method used to evaluate how important a word is to a document in a collection (corpus).

- Words like "the" , "is", "and" appear everywhere -> not useful

TF-IDF helps highlighting important words while down-weighting common ones.

TF-IDF = Term Frequency(TF) x Inverse Document Frequency (IDF)

<code>TF</code> = No. of rep of words in sentence / no. of words in sentence

AI is the future and AI is powerful.

TF("AI") = 2/8

<code>IDF</code> = LOGe (no. of sentences/ no. of sentences containg the word)

- S1 : Good boy
- S2 : Good girl
- S3 : boy girl good

```python
from sklearn.feature_extraction.text import TfidfVectorizer

documents = [
    "Good boy",
    "Good girl",
    "boy girl Good"
]

vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(documents)

print(vectorizer.get_feature_names_out())
print(X.toarray())
```
