<h1><strong>Lemmatization</strong></h1>

Lemmatization technique is like stemming, The output we get after lemmatization is called <code>lemma</code> , Which is root word rather than stem.

run, runs, running, ran 

<strong>After lemmatization</strong>

run, run, run, run

<h3>Helps in:</h3>

- Better text analysis
- Reduced vocabulary size
- better search results
- better sentiment analysis

<h3>Lemmatization vs Stemming</h3>

| Feature | Stemming | Lemmatization |
| ------|-----------|----------------|
|Method | Cuts suffix | Uses Dictionary|
|Output | May be invalid | Real world|
|Example running| run | run|
|Example Studies | studi|study|
|Accuracy|Lower|Higher|
|Speed | Fast|Slow|

<h3>How Lemmatization works</h3>

- Step1 : Identify Parts of Speech(POS)
- Step2 : Apply language rule
- Step3 : Return Dictionary Base word

<h3>Sentence</h3>

The boys are playing games and ate apples.


The boy are play and eat apple.

```python
import nltk
from nltk.stem import WordNetLemmatizer
```

```python
nltk.download('wordnet')
nltk.download('omw-1.4')
```

```python
lemmatizer = WordNetLemmatizer()
```

```python
print(lemmatizer.lemmatize("running"))
print(lemmatizer.lemmatize("studies"))
print(lemmatizer.lemmatize("boys"))
print(lemmatizer.lemmatize("writes"))
```

<h3>POS- Based Lemmatization </h3>

```python
print(lemmatizer.lemmatize("running",pos='v'))
print(lemmatizer.lemmatize("studies",pos='v'))
print(lemmatizer.lemmatize("wrote",pos='v'))
```

<h3>POS Code</h3>

n = noun
v = verb
a = adjective
r = adverb

<h3>Full sentence Lemmatization</h3>

```python
from nltk.tokenize import word_tokenize

sentence = "The boys are playing games"

words = word_tokenize(sentence)

lemmatized = [lemmatizer.lemmatize(word,pos='v') for word in words]
lemmatized
```

<h1><strong>Stopwords</strong></h1>

Stopwords are common words in language that are usually removed during text processing because they don't add much meaning full information.

This coffee is good.

- coffee good


The movie was very good and the acting was excellent.

- movie good acting excellent


Example : Articles, Pronoungs, helping verbs, prepositions, conjunctions, common adverbs

<h3><strong>Why remove stop words?</strong></h3>

- Reduce noise
- Smaller vocabulary
- Faster processing
- Better keyword extraction

<h3><strong>When not to remove stop words</strong></h3>

- Sentiment analysis
    - Ex: I do not like this movie
        - like movie

- Question Answering
    - Ex: What is the capital of india?
        - capital india

- Translation
    - Grammar words are important.


- Chatbot/Conversation AI

```python
from nltk.corpus import wordnet

def get_wordnet_pos(tag):
    if tag.startswith('J'):
        return wordnet.ADJ
    elif tag.startswith('V'):
        return wordnet.VERB
    elif tag.startswith('N'):
        return wordnet.NOUN
    elif tag.startswith('R'):
        return wordnet.ADJ
    else:
        return wordnet.NOUN
```

```python
from nltk.stem import WordNetLemmatizer

lemmatizer = WordNetLemmatizer()
speech_refined = []
for i in range(len(tagged)):
    word = tagged[i][0]
    pos = get_wordnet_pos(tagged[i][1])
    speech_refined.append(lemmatizer.lemmatize(word,pos=pos))
```
